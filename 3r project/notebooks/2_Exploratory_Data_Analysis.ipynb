# 2. Exploratory Data Analysis
Here I will explore the data visually to find insights.

```
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```
df = pd.read_csv('../data/processed/uac_metrics.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

### Time-Series Analysis
Let's look at how the system load changes over time.

```
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Date'], y=df['Children in CBP custody'], name='CBP Custody'))
fig.add_trace(go.Scatter(x=df['Date'], y=df['Children in HHS Care'], name='HHS Care'))
fig.update_layout(title='System Load Over Time')
fig.show()
```

### Flow Analysis
Comparing apprehensions vs discharges.

```
fig = go.Figure()
fig.add_trace(go.Bar(x=df['Date'], y=df['Children apprehended and placed in CBP custody'], name='Apprehensions'))
fig.add_trace(go.Bar(x=df['Date'], y=df['Children discharged from HHS Care'], name='Discharges'))
fig.update_layout(title='Daily Apprehensions & Discharges', barmode='group')
fig.show()
```

### Capacity Pressure
Checking the rolling average of Net Daily Intake.

```
fig = px.line(df, x='Date', y='Net Daily Intake (7-day avg)', title='7-Day Avg Net Intake')
fig.add_hline(y=0, line_dash="dash", line_color="red")
fig.show()
```

### Pipeline Efficiency
Looking at the efficiency ratios.

```
fig = px.line(df, x='Date', y=['CBP Efficiency Ratio', 'Discharge Capacity Ratio'], title='Efficiency Ratios (%)')
fig.add_hline(y=50, line_dash="dash", line_color="red")
fig.show()
```
